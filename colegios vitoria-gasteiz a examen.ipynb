## Los colegios a examen: un análisis de las competencias educativas modernas en los centros de Vitoria-Gasteiz.
    
Tenemos 9 valores pedagogicos indicados como que explicitamente aparecen en la página web de los centros educativos en Vitoria-Gasteiz con 1, o si negativo con 0. La suma total de estos valores es la medida de calidad que se expresa en una escala de 10, es decir un nueve puntuaría como un 10. Así mismo se ha puntuado cualitativamente este número como <5 suspenso, 5-5.9 suficiente, 6-6.9 bien, 7-8.9 notable, y >9 sobresaliente para facilitar entender esa puntuación.
    
Los datos:

```r
#organizando lo datos
datos<-read.csv("Hoja de Colegios - cop_simpl.csv",dec = ",")
#transformar tipo y valores religiosos a factor
#esto facilita algunos analisis
datos$Tipo<-as.factor(datos$Tipo)
datos$Rel<-as.factor(datos$Rel)
datos <- datos[order(datos$PUNT,decreasing = TRUE),]#de mayor a menor nota
datos
```

Organizemos los colegios de puntuación de mayor a menor en una gráfica de barras:

```r
library(ggplot2)
#ordenados por nota
ggplot(data = datos, mapping = aes(x = reorder(Centro, PUNT), PUNT),fill=factor(Tipo)) + 
  geom_bar(stat = "identity",col = "black", size = 0.25, width=0.5) + coord_flip()+theme_bw()+
    ylab("puntuación") +xlab("")
```

**Figura 1**. Gráfico de barras indicando la puntuación de los centros de enseñanza de Vitoria-Gasteiz

Solo dos colegios tienen notable, ninguno sobresaliente, veamos esto en más detalle:

```r
table(datos$Tipo,datos$NOTA)
```

Solo un centro concertado y uno cooperativo llegan a tener una nota alta y 4 concertados y 1 publico aprueban.

```r
table(datos$Rel,datos$NOTA)
#solo los laicos llegan a tener una nota alta
```

El plurilinguismo y los valores deportivos son los items que más se repiten positivamente en las webs de los centros, mientras que la presencia de alternativas a examenes tradicionales solo se defienden en un centro explicitamente.

Para probar estadísticamente si los centros escolares tienen suspenso o aprueban vamos a hacer una prueba de una muestra. Un test de una muestra examina si la media de la población es estadísticamente diferente un valore teórico, en nuestro caso un 5, que indica que un centro aprueba. Vamos a ver la normalidad de la medida puntuación:

```r
# overlay histogram and normal density
a<-ggplot(datos, aes(PUNT)) +
  stat_function(
    fun = dnorm, 
    args = list(mean = mean(datos$PUNT), sd = sd(datos$PUNT)), 
    lwd = 1, 
    col = 'black'
  )+theme_bw()+
    ylab("densidad") + 
    xlab("puntuación") + 
    ggtitle("a) Densidad")
b<-ggplot(datos, aes(PUNT)) +
  geom_histogram()+theme_bw()+
    ylab("frecuencia") + 
    xlab("puntuación") + 
    ggtitle("b) Histograma")
library(gridExtra)
grid.arrange(a,b,ncol=2)
```

**Figura 3**. Gráfico de densidad (a) e histograma (b) de la puntuaciones de los centros de Vitoria-Gasteiz en este estudio.

```r
shapiro.test(log(datos$PUNT))#no normality as data is skewed to the right
```

La medida de la muestra tiene una distribución asimétrica y no tiene normalidad (W = 0.9, p < 0.005), por lo que tenemos que usar el formato no paramétrico del test: la prueba de rango de Wilcoxon de una muestra. Esta prueba es una alternativa no paramétrica a la prueba Student t de una muestra cuando no se puede suponer que los datos estén distribuidos normalmente. Se utiliza para determinar si la mediana de la muestra es igual a un valor estándar decidido (es decir, en nuestro caso el valor teórico de aprobado = 5).

```r
median(datos$PUNT)
wilcox.colegios<-wilcox.test(datos$PUNT,mu=5,exact = F)
wilcox.colegios
#http://www.sthda.com/english/wiki/one-sample-wilcoxon-signed-rank-test-in-r
```

El tamaño del efecto (r) se calcula dividiendo el estadístico Z por la raiz cuadrada del número de observaciones (Tomczak & Tomczak, 2014)

Maciej Tomczak and Ewa Tomczak. The need to report effect size estimates revisited. An overview of some recommended measures of effect size. Trends in Sport Sciences. 2014; 1(21):19-25.

```r
library(rcompanion)
wilcoxonOneSampleR(x = datos$PUNT, mu=5)#tamaño del efecto alto
```

La puntuación de los centros (mediana = 2.2) es significativamente más pequeña que un aprobado (V = 143, p < 0.001) con un tamaño de efecto alto (r = 0.65).

Veamos ahora como cambia la puntuación dependiendo 1) del tipo de centro y dependiendo de 2) de si el centro es laico o tiene valores religiosos. Como ya hemos visto que los datos no son normales, así que para la primera pregunta usaremos una prueba de ANOVA no paramétrico: Kruskal Wallis ANOVA, y para la segunda pregunta una prueba de rango de Wilcoxon para dos muestras independientes. Dibujemos primero estas comparaciones para verlo visualmente:

```r
c<-ggplot(datos,aes(x=Tipo,y=PUNT))+geom_boxplot(fill="grey",outlier.shape = NA)+theme_bw()+
stat_summary(fun="mean",shape=7,size=3)+geom_jitter(width=0.05)+
    ylab("puntuación") +ggtitle("a) Tipo de centro")+xlab("")
```

```r
d<-ggplot(datos,aes(x=Rel,y=PUNT))+geom_boxplot(fill="grey",outlier.shape = NA)+theme_bw()+
  stat_summary(fun="mean")+ylab("puntuación") + 
    ggtitle("b) Valores religiosos")+
stat_summary(fun="mean",shape=7,size=3)+geom_jitter(width=0.05)+xlab("")
```

**Figura 4**. Gráfico de caja de las puntuaciones de los centros de Vitoria-Gasteiz por tipo (a) y por presencia de valores religiosos (b). El límite de la caja más cercano a cero indica el percentil 25, la línea horizontal línea dentro marca la mediana (la media está indicada con una cruz en un cuadro) y el límite del cuadro más alejado de cero indica el percentil 75. Los “bigotes” (líneas verticales) por encima y por debajo de la caja indican los percentiles 10 y 90. Los puntos arriba y abajo de los bigotes indican valores atípicos fuera de los percentiles 10 y 90. 

```r
kruskal.test(PUNT~Tipo,data=datos)
```

A partir del resultado de la prueba de Kruskal-Wallis, sabemos que hay una diferencia significativa entre los tipos de centro, pero no sabemos qué pares de grupos son diferentes. Una prueba de Kruskal-Wallis significativa se sigue con la prueba poshoc de Dunn para identificar estas diferencias:

```r
library(rstatix)
dunn_test(PUNT~Tipo,data=datos, p.adjust.method = "bonferroni")
```

El Eta-cuadrado se puede utilizar como medida de tamaño de efecto de la prueba de Kruskal-Wallis

eta2 = (H - k + 1)/(n - k)

donde H es el valor obtenido en la prueba Kruskal-Wallis test; k es el número de muestras; y n es el número total de observaciones (M. T. Tomczak and Tomczak 2014). Los valores de interpretación de eta-cuadrado son: 0.01- <0.06 (efecto pequeño), 0.06 - <0.14 (efecto moderado) > = 0.14 (efecto grande):

```r
kruskal_effsize(PUNT~Tipo,data=datos)
```

```r
library("dplyr")
group_by(datos, Tipo) %>%
  summarise(
    count = n(),
    mean = mean(PUNT, na.rm = TRUE),
    median = median(PUNT, na.rm = TRUE),
    sd = sd(PUNT, na.rm = TRUE),
    se =  sd(PUNT, na.rm = TRUE)/sqrt(n()) 
    )
```

Hubo diferencias estadísticamente significativas en la puntuación entre los tipos de centro evaluados mediante la prueba de Kruskal-Wallis (H = 17.784, p < 0.001), esta diferencia tuvo un tamaño de efecto grande (eta-cuadrado = 0.36). La prueba de poshoc de Dunn por pares entre centros mostró que los centros concertados (mediana = 4.4) tuvieron una puntuación más alta que los públicos (mediana = 2.2, p < 0.005), y que los centros cooperativos (mediana = 4.4) tuvieron también puntuaciones más altas que los públicos (p < 0.005), no existiendo diferencias estadísticas entre concertados y cooperativos (p = 0.46).

```r
wilcox.test(PUNT~Rel,data=datos,exact = F)
wilcox_effsize(PUNT~Rel,data=datos)
#The effect size r is calculated as Z statistic divided by square root of the sample size
```

```r
group_by(datos, Rel) %>%
  summarise(
    count = n(),
    mean = mean(PUNT, na.rm = TRUE),
    median = median(PUNT, na.rm = TRUE),
    sd = sd(PUNT, na.rm = TRUE),
    se =  sd(PUNT, na.rm = TRUE)/sqrt(n()) 
    )
```

Los centros que subrayan valores religiosos tienen puntuaciones más altas (mediana = 4.4) que los centros laicos (mediana = 2.2, W = 121, p < 0.05). Esta diferencia tiene un tamaño de efecto moderada (r = 0.34).
